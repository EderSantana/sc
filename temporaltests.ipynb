#### Reconstructions

```python
X = tensor.tensor3('features')
Z = states.apply(inputs=X)[1]
Y = tensor.dot(Z, states.proto.W)
F = function([X], Y)
E = function([X], tensor.sqr(X-Y).sum())

total_error = 0
total_energy = 0
for x_val in teststream.get_epoch_iterator():
    total_error += E(x_val[0])
    total_energy += (x_val[0]**2).sum()
```
